```python
import pandas as pd 
import numpy as np
```

```python
red_df = pd.read_csv("red_wine.csv.txt",sep=";")
white_df = pd.read_csv("white_wine.csv.txt",sep=";")
```

```python
red_df.head()
```

```python
white_df.head()
```

```python
# create color array for red dataframe
color_red = np.repeat("red",red_df.shape[0])

# create color array for white dataframe
color_white = np.repeat("white",white_df.shape[0])
```

```python
#creating color column in each dataframe
red_df["color"] = color_red
white_df["color"] = color_white
```

```python
red_df.head()
```

```python
#combine the two dataframs into one
wine_df = red_df.append(white_df,ignore_index =True,sort=False)
```

```python
wine_df.head()
```

```python
#Save Combined Dataset
wine_df.to_csv('winequality_edited.csv')
```

```python
wine_df = pd.read_csv("winequality_edited.csv")
```

```python
#rename the wrong column name
red_df= red_df.rename(columns={'total_sulfur-dioxide' : 'total_sulfur_dioxide'})
#appending again
wine_df = red_df.append(white_df,ignore_index =True,sort=False)
#saving again
wine_df.to_csv('winequality_edited.csv')
```
