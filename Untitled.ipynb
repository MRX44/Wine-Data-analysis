```python
#Q2: What level of acidity (pH value) receives the highest average rating?
""" create a categorical column from quantitative values of pH level
    Acidity Levels:
    High: Lowest 25% of pH values
    Moderately High: 25% - 50% of pH values
    Medium: 50% - 75% of pH values
    Low: 75% - max pH value
"""
```

```python
wine_df.describe()["pH"]
```

```python
# Bin edges that will be used to "cut" the data into groups
bin_edges = [2.7 ,3.11,3.21,3.32,4.01]
bin_names = ['high', 'mod_high', 'medium', 'low'] 
wine_df["acidity_levels"] = pd.cut(wine_df["pH"],bin_edges,labels = bin_names)
```

```python
wine_df.head()
```

```python
wine_df.shape
```

```python
#to answer Q2
wine_df.groupby("acidity_levels")["quality"].mean()
```

```python
#Q3: Do wines with higher alcoholic content receive better ratings?
# get the median amount of alcohol content
med = wine_df["alcohol"].median()
```

```python
# select samples with alcohol content less than the median
low_alcohol = wine_df[wine_df["alcohol"]<med]

# select samples with alcohol content greater than or equal to the median
high_alcohol = wine_df[wine_df["alcohol"]>=med]
```

```python
print(low_alcohol["quality"].mean(),high_alcohol["quality"].mean());
```

```python
#plot the results 
mean_quality_low = low_alcohol["quality"].mean()
mean_quality_high = high_alcohol["quality"].mean()
# Create a bar chart with proper labels
locations = [1, 2]
colors = ['red', 'blue']
heights = [mean_quality_low, mean_quality_high]
labels = ['Low', 'High']
plt.bar(locations, heights, tick_label=labels,color=colors,alpha=.7)
plt.title('Average Quality Ratings by Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Average Quality Rating');
```

```python
#Q4 : Do sweeter wines (more residual sugar) receive better ratings?
# get the median amount of residual sugar
med2 = wine_df["residual_sugar"].median()
```

```python
# select samples with residual sugar less than the median
low_sugar = wine_df[wine_df["residual_sugar"] < med2]

# select samples with residual sugar greater than or equal to the median
high_sugar = wine_df[wine_df["residual_sugar"] >= med2]
```

```python
print(low_sugar["quality"].mean(),high_sugar["quality"].mean())
```

```python
# Create a bar chart with proper labels
low_sugar_quality = low_sugar["quality"].mean()
high_sugar_quality = high_sugar["quality"].mean()
loc = [1,2]
colors = ['red', 'blue']
high = [low_sugar_quality,high_sugar_quality]
labels2 = ['Low Sugar', 'High Sugar']
plt.bar(loc,high,tick_label=labels2, color=colors,alpha=.7);
plt.title("Residual Sugar content")
plt.xlabel("Residual Sugar")
plt.ylabel('Average Quality Rating');
```
